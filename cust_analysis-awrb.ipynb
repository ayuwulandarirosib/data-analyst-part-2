    Then, for analysis need, we will change the 'year_birth' variable into 'age' (based on 2022) and grouped by generation to be able to know the characteristics of product purchases and the effectiveness of marketing campaigns based on age groups. Will be grouped as follows:
    a. young adult (18-35 years old)
    b. mid-age adult (36-55 years old)
    c. older adult (age >= 56 years)

    Changing 'year_birth' column into 'age' column as per 2022 and Groupping 'age' based on generation.

```python
def age_class(age):
    
    if 18 <= age <= 35:
        return 'young_adult'
    elif 36 <= age <= 55:
        return 'mid-age_adult'
    else:
        return 'older_adult'
    
ca['age'] = 2022 - ca['age']
ca.rename(columns={'age':'age'},inplace=True)

ca['age_class'] = ca['age'].apply(age_class)
ca['age_class'] = pd.Categorical(ca.age_class, categories=['young_adult','mid-age_adult','older_adult'])
```
